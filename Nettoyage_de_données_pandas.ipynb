```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\sarah\OneDrive - GROUPE 3IL\Documents\Data Analyst Bootcamp\Customer Call List.xlsx")
df
```

```python
df= df.drop_duplicates()
```

```python
df = df.drop(columns = "Not_Useful_Column")
```

```python
df["Last_Name"] = df["Last_Name"].str.lstrip("...")
df["Last_Name"] = df["Last_Name"].str.lstrip("/")
df["Last_Name"] = df["Last_Name"].str.rstrip("_")
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','', regex = "true")
df
```

```python
#df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
#df["Phone_Number"]= df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:9])
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','', regex = "true")
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','', regex = "true")
df
```

```python
df[["Rue","Ville","Code_postal"]] = df["Address"].str.split(',',n=2,expand=True )
df
```

```python
df["Paying Customer"]= df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"]= df["Paying Customer"].str.replace('No','N')
```

```python
df["Do_Not_Contact"]= df["Paying Customer"].str.replace('Yes','Y')
df["Do_Not_Contact"]= df["Paying Customer"].str.replace('No','N')
df
```

```python
df = df.replace('N/a','')
```

```python
df
```

```python
df = df.fillna('')
```

```python
df
```

```python
for x in df.index:
    if df.loc[x,"Phone_Number"] == '':
        df.drop(x, inplace=True)
```
